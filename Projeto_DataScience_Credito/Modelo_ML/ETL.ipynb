```python
solicitacoes_credito2 = solicitacoes_credito.drop(['numero_solicitacao', 'razaoSocial', 'nomeFantasia', 'cnpjSemTraco', 'dataAprovadoEmComite', 'scorePontualidade', 'dashboardCorrelacao', 'definicaoRisco', 'diferencaPercentualRisco', 'margemBrutaAcumulada', 'percentualProtestos', 'periodoDemonstrativoEmMeses', 'restricoes', 'empresa_MeEppMei', 'dataAprovadoNivelAnalista', 'anoFundacao', 'percentualRisco','primeiraCompra'], axis=1)

solicitacoes_credito2.head(2)
solicitacoes_credito2.info()
```

```python
solicitacoes_credito3 = solicitacoes_credito2

solicitacoes_credito3 = solicitacoes_credito3.drop_duplicates()

# solicitacoes_credito3['ano_primeiraCompra'] = pd.to_datetime(solicitacoes_credito3['primeiraCompra'], errors='coerce').dt.year

# solicitacoes_credito3['primeiraCompra_valida'] = solicitacoes_credito3['primeiraCompra'].notna().astype(int)

# solicitacoes_credito3 = solicitacoes_credito3.drop(['primeiraCompra'], axis=1)

solicitacoes_credito3 = solicitacoes_credito3.dropna(subset=['totalPatrimonioLiquido'])

max_atraso = solicitacoes_credito3['maiorAtraso'].max()

solicitacoes_credito3['score_atraso'] = 1 - (solicitacoes_credito3['maiorAtraso'] / max_atraso)

# Garantir que o score esteja no intervalo [0, 1]
solicitacoes_credito3['score_atraso'] = np.clip(solicitacoes_credito3['score_atraso'], 0, 1)

solicitacoes_credito3 = solicitacoes_credito3[solicitacoes_credito3['totalPatrimonioLiquido'].notnull()]
# solicitacoes_credito3 = solicitacoes_credito3[solicitacoes_credito3['totalPatrimonioLiquido'] != 0]

solicitacoes_credito3 = solicitacoes_credito3.fillna(0)

solicitacoes_credito3.info()
```

```python
# Inicializar o scaler
scaler = MinMaxScaler()

# Normalizar cada componente
solicitacoes_credito3[['endividamento', 'titulosEmAberto', 'prazoMedioRecebimentoVendas', 'score_atraso', 'margemBruta', 'valorAprovado', 'valorSolicitado']] = scaler.fit_transform(
    solicitacoes_credito3[['endividamento', 'titulosEmAberto', 'prazoMedioRecebimentoVendas', 'score_atraso', 'margemBruta', 'valorAprovado', 'valorSolicitado']]
)
```

```python
# Função para calcular o risco ponderado com base nos fatores informados
def calcular_risco(row):
    # Pesos atribuídos a cada fator de risco
    pesos = {
        'maiorAtraso': 0.2,  # Maior peso ao histórico de atraso
        'prazoMedioRecebimentoVendas': 0.1,  # Fluxo de caixa importante, mas não crítico
        'titulosEmAberto': 0.1,  # Dívidas pendentes são um risco importante
        'valorSolicitado': 0.05,  # Solicitado tem peso menor que o valor aprovado
        'valorAprovado': 0.05,
        'ativoCirculante': 0.05,
        'passivoCirculante': 0.05,
        'endividamento': 0.1,  # Endividamento tem um peso significativo
        'faturamentoBruto': 0.1,
        'margemBruta': 0.05,
        'capitalSocial': 0.05,
        'limiteEmpresaAnaliseCredito': 0.05,
        'score_atraso': 0.1,  # Score de atraso é relevante
    }
    
    # Normalizando algumas variáveis para ficarem em escalas semelhantes
    ativo_liquidez = row['ativoCirculante'] / (row['passivoCirculante'] + 1)  # Evitar divisão por zero
    valor_solicitado_aprovado = row['valorSolicitado'] / (row['valorAprovado'] + 1)
    
    # Cálculo do score ponderado
    risco = (
        pesos['maiorAtraso'] * row['maiorAtraso'] +
        pesos['prazoMedioRecebimentoVendas'] * row['prazoMedioRecebimentoVendas'] +
        pesos['titulosEmAberto'] * row['titulosEmAberto'] +
        pesos['valorSolicitado'] * valor_solicitado_aprovado +
        pesos['ativoCirculante'] * ativo_liquidez +
        pesos['endividamento'] * row['endividamento'] +
        pesos['faturamentoBruto'] * row['faturamentoBruto'] +
        pesos['margemBruta'] * row['margemBruta'] +
        pesos['capitalSocial'] * row['capitalSocial'] +
        pesos['limiteEmpresaAnaliseCredito'] * row['limiteEmpresaAnaliseCredito'] +
        pesos['score_atraso'] * (1 - row['score_atraso'])  # Quanto maior o score, menor o risco
    )
    
    # Retornar o valor de risco calculado
    return risco

# Aplicando a função de cálculo de risco para cada linha do DataFrame
solicitacoes_credito3['risco_calculado'] = solicitacoes_credito3.apply(calcular_risco, axis=1)

# Normalizando os valores de risco entre 0 e 1
solicitacoes_credito3['risco_calculado'] = np.clip(solicitacoes_credito3['risco_calculado'], 0, 1)

# Exibindo os resultados com o risco calculado
solicitacoes_credito3[['maiorAtraso', 'prazoMedioRecebimentoVendas', 'titulosEmAberto', 'valorSolicitado', 
    'valorAprovado', 'ativoCirculante', 'passivoCirculante', 'endividamento', 
    'faturamentoBruto', 'margemBruta', 'capitalSocial', 'limiteEmpresaAnaliseCredito', 
    'score_atraso', 'risco_calculado']].head()

```
