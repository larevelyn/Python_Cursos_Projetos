```python
vendas_dez_df = pd.read_excel("Arquivos/Vendas - Dez.xlsx")
display(vendas_dez_df)
```

```python
# adicionar linhas/empilhar bases
# vendas_df = vendas_df.append(vendas_dez_df)
vendas_df = pd.concat([vendas_df, vendas_dez_df], ignore_index=True)
display(vendas_df)

# Remover a coluna 'data'
vendas_df = vendas_df.drop(columns=['Data'])
```

```python
# # # excluir linhas e colunas
# # axis 1 é eixo da coluna, e axis 0 é da linha

# # colunas
vendas_df = vendas_df.drop("Imposto", axis=1)
# linhas
vendas_df = vendas_df.drop(0, axis=0)

display(vendas_df)
```

```python
# Valores Vazios 
# Deletar linhas e colunas vazias totalmente vazias
vendas_df = vendas_df.dropna(how='all')
# excluir colunas vazias
vendas_df = vendas_df.dropna(how='all' axis=1)
# exluir linhas que tem pelo menos 1 valor vazio
vendas_df = vendas_df.dropna()
# excluir linhas onde coluna X for vazia
vendas_df = vendas_df.dropna(subset=['Comissão'])
```

```python
# preencher valores vazios
# peencher os valores nulos com 0
# vendas_df['Comissão'] = vendas_df['Comissão'].fillna(0)
vendas_df2 = vendas_df['Comissão'].fillna(0)

# preencher com a média da coluna comissão
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())

display(vendas_df)
```

```python
# preencher nulos com o ultimo valor, ou seja, com o valor da linha de cima
vendas_df = vendas_df.ffill()

display(vendas_df)
```

```python
# value counts - count do sql, parecido com select id_loja, count(id_loja) from vendas_df
transacoes_lojas = vendas_df['ID Loja'].value_counts()
display(transacoes_lojas)
```

```python
# groupby - serve agrupar produtos e somar (como no sql)
# Nessa linha ele vai agrupar por Produto e somar as outras linhas que são numericas
faturamento_produto1 = vendas_df.groupby('Produto').sum()
# Aqui ele faz a mesma coisa da linha de cima, porém ele só vai trazer produto e valor final
# pode ser mean(media), max, min e etc
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

```python
gerentes_df = pd.read_excel('Arquivos/Gerentes.xlsx')
display(gerentes_df)
```

```python
# Merge - mesclar duas tabelas - Join
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```

```python
# # Conta quantos nulos tem na coluna
# vendas_df['Quantidade'].isna().sum()

# Substitui valores nulos pela média
vendas_df.loc[pd.isnull(vendas_df.Quantidade), 'Quantidade'] = vendas_df.Quantidade.mean() 

# Substitui valores nulos por 0
vendas_df.loc[pd.isnull(vendas_df.Produto), 'Produto'] = 0  

# Mantém os valores iguais (sem mudança real)
vendas_df['Produto'] = vendas_df.Produto.Produto('Mochila', 'Mochila')  

# Substitui 0 por 'no' 
vendas_df['Produto'] = vendas_df.Produto.replace(0, 'no') 

# Substitui 'o' por 'no'
vendas_df['Produto'] = vendas_df.Produto.replace('o', 'no')  

# Ajusta capitalização - Deixa todos os campos que começam com minusculo com maiusculo sem mudar os corretos
vendas_df['Gerente'] = vendas_df.Gerente.replace('Camiseta', 'camiseta').replace('Relógio', 'relógio').replace('Chinelo Liso', 'chinelo liso')  
```

```python
# transformar todos os dados não numericos em numericos fazendo que as linhas virem colunas
# por exemplo, ao invés de ter uma colunas gerentes, eu tenho uma coluna para cada gerente com 0 e 1 ou true e false
vendas_dummies = pd.get_dummies(vendas_df)
vendas_dummies.head()
```

```python
# mostra a quantidade de nulos por coluna
nulos_por_coluna = vendas_df.isnull().sum()
print(nulos_por_coluna)
```

```python
# Filtrar para manter apenas colunas com nulos diferentes de zero
colunas_com_nulos = nulos_por_coluna[nulos_por_coluna > 0]
print(colunas_com_nulos)
```
